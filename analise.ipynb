```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('brasileirao.csv')
```

```python
df.head(24)
```

```python
df2024 = df.loc[df['season'] == 2024]
```

```python
df2024
```

```python
df2024.reset_index()
```

```python
df2024 = df2024.reset_index()
```

```python
df2024 = df2024.drop(columns = 'index')
```

```python
df2024['Classif_Libertadores'] = (df2024['place'] > 0) & (df2024['place'] <= 4)
```

```python
df2024
```

```python
df2024
```

```python
df2024['Classif_Pré'] = (df2024['place'] >= 5) & (df2024['place'] <= 6)
```

```python
df2024
```

```python
df2024['Rebaixado'] = (df2024['place'] >= 17) & (df2024['place'] <= 20)
```

```python
df2024
```

```python
dftop3 = (df2024['place'] >=1) & (df2024['place'] <= 3)
```

```python
dftop3 = df2024[['place', 'team', 'points', 'won', 'loss', 'goals']]
```

```python
dftop3
```

```python
dfTop3 = dftop3.iloc[0 : 3]
```

```python
dfTop3
```

```python
graf = dfTop3.plot(x = 'team', y = 'points', kind = 'bar', color = ['black', 'green', 'red'])
plt.title('Top 3 Brasileirão 2024')
plt.xlabel('Time')
plt.ylabel('Pontuação')
plt.xticks(rotation = 0)
graf.get_legend().remove()
graf.bar_label(graf.containers[0])
plt.grid(
    axis='y',
    linestyle='--',   # tracejado
    linewidth=0.7,
    alpha=0.7         # transparência
)
plt.savefig('graficoBR.pdf')
plt.show()
```
